```python
from src.vfm.connection import Connection
from src.vfm.utils.descriptive_utils import get_all_wells
from src.vfm.preprocessor import Preprocessor
import pandas as pd
import matplotlib.pyplot as plt
import os
```

```python
connection = Connection()
# Get client
client = connection.get_client()
```

```python
all_vars={
            "qo_well_test": "Sm3/h",
            "qg_well_test": "Sm3/h", 
            "qw_well_test": "Sm3/h",
            "qo_mpfm": "Sm3/h",
            "qg_mpfm": "Sm3/h", 
            "wc_mpfm": "%",
            "dhp": "bara", 
            "dht": "°C", 
            "whp": "bara", 
            "wht": "°C",
            "choke": "%", 
            "dcp": "bara",
            "gl_mass_rate": "kg/h",
            "gl_open_ratio": "unitless"
           }
```

```python
preprocessor = Preprocessor()
```

```python
output_dir = r"doc\4-thesis\images"
os.makedirs(output_dir, exist_ok=True)

for well in get_all_wells():
    try:
        print(f"Processing {well}")
        df = connection.get_data(client, wells=[well])

        # Save data
        df.to_csv(rf"data\{well}.csv", index=True)

        # Print stats
        min_ts = df.index.min()
        max_ts = df.index.max()
        print(well, df.shape[0], f"min_ts={min_ts}", f"max_ts={max_ts}")

        # Plot each independent variable vs time
        for col in df.columns:
            plt.figure(figsize=(10, 4))
            plt.plot(df.index, df[col],  marker="o", linestyle="")
            plt.title(f"{well} - {col} over time")
            plt.xlabel("Time")
            plt.ylabel(f"{col} ({all_vars.get(col)})")
            plt.tight_layout()

            # Save plot
            plt.savefig(os.path.join(output_dir, f"{well}_{col}.png"))
            plt.close()

    except Exception as e:
        print(e)
        pass
```
