```python
from src.vfm.connection import Connection
import pandas as pd
import matplotlib.pyplot as plt
import os
```

```python
connection = Connection()
# Get client
client = connection.get_client()
```

```python
all_vars={
            "qo_well_test": "Sm3/h",
            "qg_well_test": "Sm3/h", 
            "qw_well_test": "Sm3/h",
            "qo_mpfm": "Sm3/h",
            "qg_mpfm": "Sm3/h", 
            "wc_mpfm": "%",
            "dhp": "bara", 
            "dht": "°C", 
            "whp": "bara", 
            "wht": "°C",
            "choke": "%", 
            "dcp": "bara"
           }
```

```python
output_dir = r"doc\thesis\images"
os.makedirs(output_dir, exist_ok=True)

for i in [8]:
    try:
        well = f"W0{i}" if i < 10 else f"W{i}"
        df_raw = connection.get_data(client, well=well)

        # Save data
        df_raw.to_csv(rf"data\{well}.csv", index=True)

        # Print stats
        min_ts = df_raw.index.min()
        max_ts = df_raw.index.max()
        print(well, df_raw.shape[0], f"min_ts={min_ts}", f"max_ts={max_ts}")

        print(df_raw["whp"].describe)

        # Plot each independent variable vs time
        for col in df_raw.columns:
            plt.figure(figsize=(10, 4))
            plt.plot(df_raw.index, df_raw[col],  marker="o", linestyle="")
            plt.title(f"{well} - {col} over time")
            plt.xlabel("Time")
            plt.ylabel(f"{col} ({all_vars.get(col)})")
            plt.tight_layout()

            # Save plot
            plt.savefig(os.path.join(output_dir, f"{well}_{col}.png"))
            plt.close()

    except Exception:
        pass
```
